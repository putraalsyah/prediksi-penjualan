```python
"""
LANGKAH 1: PAHAMI KONSEP DASAR

Rangkuman Konsep Regresi Linear:

1. Apa tujuan utama Regresi Linear?
   Tujuan utama Regresi Linear adalah untuk memodelkan dan memprediksi hubungan 
   linear antara variabel independen (input) dengan variabel dependen (output), 
   sehingga kita dapat memahami seberapa besar pengaruh satu variabel terhadap 
   variabel lainnya dan melakukan prediksi nilai di masa depan.

2. Apa yang direpresentasikan oleh "garis lurus terbaik" (best-fit line)?
   Garis lurus terbaik merepresentasikan hubungan matematis optimal antara variabel 
   X dan Y yang meminimalkan total error (jarak vertikal) antara titik-titik data 
   aktual dengan garis prediksi. Garis ini dihitung menggunakan metode seperti 
   Ordinary Least Squares untuk mendapatkan parameter terbaik (slope dan intercept).
"""

print("="*80)
print("LANGKAH 2: FORMULIR PERANCANGAN PROYEK TA-02")
print("="*80)

formulir_proyek = """
1. JUDUL PROYEK:
   Analisis Pengaruh Biaya Iklan TV terhadap Penjualan Produk

2. LATAR BELAKANG MASALAH:
   Sebagai seorang marketing analyst, saya ingin mengetahui seberapa efektif 
   investasi biaya iklan di media TV dalam meningkatkan penjualan produk. Dengan 
   memahami hubungan ini, perusahaan dapat mengalokasikan budget marketing secara 
   lebih optimal dan memprediksi ROI (Return on Investment) dari kampanye iklan.

3. TAUTAN DATASET:
   https://www.kaggle.com/datasets/ashydv/advertising-dataset
   
   CARA DOWNLOAD:
   1. Kunjungi link di atas
   2. Login/Register di Kaggle (gratis)
   3. Klik tombol "Download" untuk mendapatkan file Advertising.csv
   4. Simpan file di folder yang sama dengan notebook ini
   

4. VARIABEL YANG AKAN DIGUNAKAN:
   - Variabel Independen (X): TV (budget iklan TV dalam ribuan dollar)
   - Variabel Dependen (y): Sales (penjualan produk dalam ribuan unit)

5. BUKTI AKSES DAN VALIDASI DATA:
   [Screenshot akan ditampilkan di bawah]
"""

print(formulir_proyek)
```

```python
print(formulir_proyek)

# ========================================
# BUKTI AKSES DAN VALIDASI DATA
# ========================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
import warnings
warnings.filterwarnings('ignore')

# Set style untuk visualisasi
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)

print("\n" + "="*80)
print("MEMUAT DATASET")
print("="*80)

print("\nüìå INSTRUKSI PENTING:")
print("Download dataset dari: https://www.kaggle.com/datasets/ashydv/advertising-dataset")
print("Simpan file 'Advertising.csv' di folder yang sama dengan notebook ini\n")

# Ganti 'Advertising.csv' dengan path file Anda jika berbeda
try:
    df = pd.read_csv('Advertising.csv')
    print("‚úì Dataset berhasil dimuat!")
except FileNotFoundError:
    print("‚ùå ERROR: File 'Advertising.csv' tidak ditemukan!")
    print("\nSilakan download dataset terlebih dahulu dari:")
    print("https://www.kaggle.com/datasets/ashydv/advertising-dataset")
    print("\nAtau gunakan alternatif:")
    print("https://www.kaggle.com/datasets/bumba5341/advertisingcsv")
    print("\nSetelah download, simpan file di folder yang sama dengan notebook ini.")
    raise

print("\n--- 5 BARIS PERTAMA ---")
print(df.head())

print("\n--- NAMA KOLOM ---")
print(df.columns.tolist())

print("\n--- INFORMASI DATASET ---")
print(df.info())

print("\n--- STATISTIK DESKRIPTIF ---")
print(df.describe())

print("\n--- DIMENSI DATASET ---")
print(f"Jumlah Baris: {df.shape[0]}")
print(f"Jumlah Kolom: {df.shape[1]}")

print("\n" + "="*80)
print("TP-03 SELESAI - LANJUT KE TUGAS AKHIR (TA-02)")
print("="*80)
```

```python
# ========================================
# TUGAS AKHIR (TA-02)
# Implementasi Regresi Linear pada Studi Kasus Advertising
# ========================================

print("\n" + "="*80)
print("1. EKSPLORASI DATA (EDA)")
print("="*80)

# Cek missing values
print("\n--- MISSING VALUES ---")
print(df.isnull().sum())

# Cek duplikat
print(f"\nJumlah data duplikat: {df.duplicated().sum()}")

# Korelasi antar variabel
print("\n--- MATRIKS KORELASI ---")
correlation_matrix = df.corr()
print(correlation_matrix)

# Visualisasi 1: Pairplot
print("\n[Membuat Pairplot untuk melihat hubungan antar variabel...]")
plt.figure(figsize=(12, 10))
sns.pairplot(df, diag_kind='kde', plot_kws={'alpha': 0.6})
plt.suptitle('Pairplot: Hubungan Antar Variabel Advertising', y=1.02, fontsize=16)
plt.tight_layout()
plt.savefig('pairplot_advertising.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Visualisasi 2: Heatmap Korelasi
print("\n[Membuat Heatmap Korelasi...]")
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, 
            square=True, linewidths=1, cbar_kws={"shrink": 0.8}, fmt='.3f')
plt.title('Heatmap Korelasi: Advertising Dataset', fontsize=16, pad=20)
plt.tight_layout()
plt.savefig('heatmap_correlation.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Visualisasi 3: Scatterplot TV vs Sales (Validasi Hubungan Linear)
print("\n[Membuat Scatterplot TV vs Sales untuk validasi hubungan linear...]")
plt.figure(figsize=(10, 6))
plt.scatter(df['TV'], df['Sales'], alpha=0.6, s=50, edgecolors='black', linewidth=0.5)
plt.xlabel('Budget Iklan TV (ribuan dollar)', fontsize=12)
plt.ylabel('Penjualan (ribuan unit)', fontsize=12)
plt.title('Hubungan Antara Budget Iklan TV dan Penjualan\n(Validasi Linearitas)', 
          fontsize=14, pad=15)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('scatterplot_tv_sales.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n‚úì Dari scatterplot terlihat adanya tren linear positif antara TV dan Sales")
print("‚úì Ini memvalidasi bahwa Regresi Linear cocok untuk studi kasus")
```

```python
# ========================================
# 2. PERSIAPAN DATA UNTUK MODEL
# ========================================

print("\n" + "="*80)
print("2. PERSIAPAN DATA UNTUK MODEL")
print("="*80)

# Memisahkan fitur (X) dan target (y)
# Fokus pada TV sebagai variabel independen (Regresi Linear Sederhana)
X = df[['TV']]
y = df['Sales']

print(f"\nShape X (fitur): {X.shape}")
print(f"Shape y (target): {y.shape}")

# Membagi data: 70% training, 30% testing
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

print(f"\nData Training: {X_train.shape[0]} samples")
print(f"Data Testing: {X_test.shape[0]} samples")
```

```python
# ========================================
# 3. PENGEMBANGAN MODEL REGRESI LINEAR
# ========================================

print("\n" + "="*80)
print("3. PELATIHAN MODEL REGRESI LINEAR")
print("="*80)

# Membuat instance model
model = LinearRegression()

# Melatih model
print("\n[Melatih model...]")
model.fit(X_train, y_train)
print("‚úì Model berhasil dilatih!")

# Menampilkan parameter model
print(f"\n--- PARAMETER MODEL ---")
print(f"Intercept (Œ≤‚ÇÄ): {model.intercept_:.4f}")
print(f"Coefficient TV (Œ≤‚ÇÅ): {model.coef_[0]:.4f}")

print(f"\n--- INTERPRETASI MODEL ---")
print(f"Persamaan Regresi: Sales = {model.intercept_:.4f} + {model.coef_[0]:.4f} √ó TV")
print(f"\nMakna Bisnis:")
print(f"‚Ä¢ Intercept ({model.intercept_:.4f}): Penjualan dasar tanpa iklan TV")
print(f"  (dalam ribuan unit)")
print(f"‚Ä¢ Coefficient ({model.coef_[0]:.4f}): Setiap penambahan $1000 budget iklan TV")
print(f"  akan meningkatkan penjualan sebesar {model.coef_[0]:.4f} ribu unit")
print(f"  atau sekitar {model.coef_[0]*1000:.0f} unit")
```

```python
# ========================================
# 4. PREDIKSI DAN EVALUASI MODEL
# ========================================

print("\n" + "="*80)
print("4. PREDIKSI DAN EVALUASI MODEL")
print("="*80)

# Melakukan prediksi pada data test
y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)

# Evaluasi pada data training
mae_train = metrics.mean_absolute_error(y_train, y_pred_train)
mse_train = metrics.mean_squared_error(y_train, y_pred_train)
rmse_train = np.sqrt(mse_train)
r2_train = metrics.r2_score(y_train, y_pred_train)

# Evaluasi pada data testing
mae_test = metrics.mean_absolute_error(y_test, y_pred_test)
mse_test = metrics.mean_squared_error(y_test, y_pred_test)
rmse_test = np.sqrt(mse_test)
r2_test = metrics.r2_score(y_test, y_pred_test)

print("\n--- METRIK EVALUASI DATA TRAINING ---")
print(f"Mean Absolute Error (MAE): {mae_train:.4f}")
print(f"Mean Squared Error (MSE): {mse_train:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse_train:.4f}")
print(f"R-squared (R¬≤): {r2_train:.4f}")

print("\n--- METRIK EVALUASI DATA TESTING ---")
print(f"Mean Absolute Error (MAE): {mae_test:.4f}")
print(f"Mean Squared Error (MSE): {mse_test:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse_test:.4f}")
print(f"R-squared (R¬≤): {r2_test:.4f}")

print("\n--- INTERPRETASI METRIK ---")
print(f"‚Ä¢ R¬≤ = {r2_test:.2%}: Model dapat menjelaskan {r2_test:.2%} variasi penjualan")
print(f"‚Ä¢ RMSE = {rmse_test:.2f}: Rata-rata error prediksi sekitar {rmse_test:.2f} ribu unit")
print(f"  atau {rmse_test*1000:.0f} unit")
```

```python
# Visualisasi 4: Actual vs Predicted (Training Data)
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.scatter(y_train, y_pred_train, alpha=0.6, s=50, edgecolors='black', linewidth=0.5)
plt.plot([y_train.min(), y_train.max()], [y_train.min(), y_train.max()], 
         'r--', lw=2, label='Perfect Prediction')
plt.xlabel('Penjualan Aktual (ribuan unit)', fontsize=11)
plt.ylabel('Penjualan Prediksi (ribuan unit)', fontsize=11)
plt.title(f'Training Set\nR¬≤ = {r2_train:.4f}', fontsize=12)
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
plt.scatter(y_test, y_pred_test, alpha=0.6, s=50, edgecolors='black', linewidth=0.5, color='orange')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 
         'r--', lw=2, label='Perfect Prediction')
plt.xlabel('Penjualan Aktual (ribuan unit)', fontsize=11)
plt.ylabel('Penjualan Prediksi (ribuan unit)', fontsize=11)
plt.title(f'Testing Set\nR¬≤ = {r2_test:.4f}', fontsize=12)
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('actual_vs_predicted.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Visualisasi 5: Regression Line
plt.figure(figsize=(12, 6))
plt.scatter(X_train, y_train, alpha=0.5, s=50, label='Training Data', edgecolors='black', linewidth=0.5)
plt.scatter(X_test, y_test, alpha=0.5, s=50, color='orange', label='Testing Data', edgecolors='black', linewidth=0.5)

# Plot garis regresi
X_range = np.linspace(X.min().values[0], X.max().values[0], 100).reshape(-1, 1)
y_range_pred = model.predict(X_range)
plt.plot(X_range, y_range_pred, 'r-', linewidth=3, 
         label=f'Regression Line\ny = {model.intercept_:.2f} + {model.coef_[0]:.4f}x')

plt.xlabel('Budget Iklan TV (ribuan dollar)', fontsize=12)
plt.ylabel('Penjualan (ribuan unit)', fontsize=12)
plt.title('Model Regresi Linear: TV Advertising vs Sales', fontsize=14, pad=15)
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('regression_line.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Visualisasi 6: Residual Plot
residuals = y_test - y_pred_test

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.scatter(y_pred_test, residuals, alpha=0.6, s=50, edgecolors='black', linewidth=0.5)
plt.axhline(y=0, color='r', linestyle='--', linewidth=2)
plt.xlabel('Nilai Prediksi', fontsize=11)
plt.ylabel('Residual', fontsize=11)
plt.title('Residual Plot', fontsize=12)
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
plt.hist(residuals, bins=20, edgecolor='black', alpha=0.7)
plt.xlabel('Residual', fontsize=11)
plt.ylabel('Frekuensi', fontsize=11)
plt.title('Distribusi Residual', fontsize=12)
plt.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.savefig('residual_analysis.png', dpi=300, bbox_inches='tight')
plt.show()
```
